# Preprocessing before sentiment analysis

```python
def normalize_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = ''.join(char for char in text if char.isalpha() or char.isspace())
    return text
```

```python
balanced_df['Title'] = balanced_df['Title'].apply(normalize_text)
```

```python
balanced_df.head()
```

```python
balanced_df['tokens'] = balanced_df['Title'].apply(lambda x: x.split())
# splitting the title in individual words
```

```python
balanced_df.head()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
balanced_df['tokens'] = balanced_df['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
# remove stop words
```

```python
balanced_df.head()
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
balanced_df['tokens'] = balanced_df['tokens'].apply(lambda x: [stemmer.stem(word) for word in x])
# making words in base form
```

```python
balanced_df.head()
```

```python
balanced_df['processed_title'] = balanced_df['tokens'].apply(lambda x: ' '.join(x))
# joining back to make one title sentence
```

# LSTM RNN 

```python
texts1 = balanced_df['processed_title'].values # Column containing news articles
labels1 = balanced_df['Sentiment'].values
```

```python
label_encoder = LabelEncoder()
encoded_labels1 = label_encoder.fit_transform(labels1)
```

```python
num_classes1 = len(np.unique(encoded_labels1))
categorical_labels1 = to_categorical(encoded_labels1, num_classes1)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(texts1, categorical_labels1, test_size=0.2, random_state=42)
```

```python
X_train2.shape
```

```python
y_train2.shape
```

```python
from collections import Counter
```

```python
all_text = " ".join(texts1)
unique_words = set(all_text.split())
len(unique_words)
```

```python
max_words = len(unique_words)  # Maximum vocabulary size
max_len = 100  # Maximum sequence length
```

```python
max_words
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
word_tokenizer = Tokenizer()
word_tokenizer.fit_on_texts(X_train2)

X_train_seq = word_tokenizer.texts_to_sequences(X_train2)
X_test_seq = word_tokenizer.texts_to_sequences(X_test2)
```

```python
X_train_seq
```

```python
# tokenizer = Tokenizer(num_words=max_words)
# tokenizer.fit_on_texts(X_train2)
```

```python
# X_train_seq = tokenizer.texts_to_sequences(X_train2)
# X_test_seq = tokenizer.texts_to_sequences(X_test2)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')
```

```python
X_train_pad.shape
```

```python
X_train_pad.shape
```

```python
from numpy import asarray
from numpy import zeros

embeddings_dictionary = dict()
glove_file = open('glove.6B.100d.txt', encoding="utf8")

for line in glove_file:
    records = line.split()
    word = records[0]
    vector_dimensions = asarray(records[1:], dtype='float32')
    embeddings_dictionary [word] = vector_dimensions
glove_file.close()
```

```python
vocab_length = len(word_tokenizer.word_index) + 1

vocab_length
```

```python
embedding_matrix = zeros((vocab_length, 100))
for word, index in word_tokenizer.word_index.items():
    embedding_vector = embeddings_dictionary.get(word)
    if embedding_vector is not None:
        embedding_matrix[index] = embedding_vector
```

```python
embedding_matrix.shape
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
```

```python
lstm_model = Sequential()
embedding_layer = Embedding(vocab_length, 100, weights=[embedding_matrix], input_length=max_len , trainable=False)

lstm_model.add(embedding_layer)
lstm_model.add(LSTM(128))
lstm_model.add(Dense(num_classes, activation='softmax'))
```

```python
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
print(lstm_model.summary())
```

```python
y_train2.shape
```

```python
lstm_model_history = lstm_model.fit(X_train_pad, y_train2, batch_size=32, epochs=10)
```

```python
score = lstm_model.evaluate(X_test_pad, y_test2, verbose=1)
```

```python
print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
