```python
df1=df['Close']
```

```python
df1=pd.DataFrame(df1)
```

```python
df1
```

```python
import pandas_ta as ta
```

```python
df1['SMA_20'] = ta.sma(df1['Close'], length=20)        
df1['EMA_20'] = ta.ema(df1['Close'], length=20)        
df1['RSI'] = ta.rsi(df1['Close'], length=14)           
df1['MACD'] = ta.macd(df1['Close']).iloc[:, 0]         
df1['MACD_signal'] = ta.macd(df1['Close']).iloc[:, 1]
```

```python
df1.head(50)
```

```python
df1.dropna(inplace=True)
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1=df1.tail(30)
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1.tail()
```

```python
import joblib
```

```python
scaler_close=joblib.load('Scale_closingPrice.pkl')
```

```python
scaler_indicators=joblib.load('Scale_Indicator.pkl')
```

```python
scaled_close = scaler_close.transform(df1[['Close']])
scaled_indicators = scaler_indicators.transform(df1[['SMA_20', 'EMA_20', 'RSI', 'MACD', 'MACD_signal']])
```

```python
scaled_close.shape
```

```python
scaled_indicators.shape
```

```python
scaled_data = np.hstack((scaled_close, scaled_indicators))
```

```python
scaled_data.shape
```

```python
# Prepare the time series data
window_size = 30

# Since you only have 30 days of data, X will just be one sequence of the last 30 days
X = scaled_data[-window_size:] 

# Reshape X to fit the model's expected input format for LSTM: (samples, time steps, features)
X = np.expand_dims(X, axis=0)
```

```python
X.shape
```

```python
from tensorflow.keras.models import load_model
from keras.metrics import mean_squared_error
```

```python
model = load_model('lstm_model2.h5', custom_objects={'mse': mean_squared_error})
```

```python
prediction = model.predict(X).flatten()
```

```python
prediction
```

```python
actual_prediction=scaler_close.inverse_transform(prediction.reshape(-1, 1))
```
